```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# read data
baseDir = '../data/sampleImpact'
plotDir = '../plots'
x = np.loadtxt(baseDir+'/x1.txt')
F = np.loadtxt(baseDir+'/Fx.txt')
g = np.loadtxt(baseDir+'/gx.txt')
```

```python
#
fig = plt.figure(figsize=[8,6])
ax = plt.subplot(311)
plt.plot(x[...,0]-50,x[...,1],label='S/C displacement')
plt.grid(True)
plt.ylabel('nm',fontsize=14)
ax.set_xticklabels([])
ax.tick_params(labelsize=14)
ax.legend(loc=4,fontsize=14)
ax2 = plt.subplot(312)
plt.plot(F[...,0]-50,-F[...,1],color = 'tab:orange',label='S/C force')
plt.ylabel('$\mu$N',fontsize=14)
plt.grid(True)
ax2.set_xticklabels([])
ax2.tick_params(labelsize=14)
ax2.legend(loc=1,fontsize=14)
ax3 = plt.subplot(313)
plt.plot(g[...,0]-50,g[...,1],color = 'tab:green',label='S/C accel')
plt.ylabel('nm s$^{-2}$',fontsize=14)
plt.xlabel('seconds from 2018-07-31 18:18:48 UTC',fontsize=14)
plt.grid(True)
ax3.tick_params(labelsize=14)
ax3.legend(loc=4,fontsize=14)
fig.tight_layout()
fig.savefig(plotDir + '/sampleTimeseries.eps', format = 'eps')
```
